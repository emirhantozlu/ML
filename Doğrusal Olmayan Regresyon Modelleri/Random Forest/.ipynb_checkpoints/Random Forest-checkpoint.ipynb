### Topluluk Öğrenme Yöntemleri

Topluluk öğrenme yöntemleri, birden fazla algoritma veya modelin bir araya gelerek ortak bir tahmin veya sınıflandırma yapmasını amaçlayan yöntemlerdir. Bu yöntemler, genellikle bireysel modellerin performansını artırmak ve daha sağlam tahminler elde etmek için kullanılır.

### Bootstrap Rastgele Örnekleme Yöntemi

**Bootstrap rastgele örnekleme yöntemi**, veri setinden yerine koymalı olarak tekrar tekrar örnekler almayı ifade eder. Bu yöntem, farklı modellerin aynı veri setinden farklı örnekler üzerinde eğitilmesine olanak tanır ve topluluk yöntemlerinin temelinde yer alır.

### Bagging (Bootstrap Aggregating)

Bagging, birden fazla karar ağacı gibi modellerin bootstrap örnekleme yöntemiyle eğitilmesi ve bu modellerin tahminlerinin birleştirilerek daha iyi bir genel tahmin yapılması prensibine dayanır.

**Bagging’in Avantajları:**

- **Hata Kareler Ortalamasının Karekökü (RMSE) Değerini Düşürür**: Modelin hata oranını azaltır.
- **Doğru Sınıflandırma Oranını Artırır**: Tahminlerin doğruluğunu artırır.
- **Varyansı Düşürür**: Modelin aşırı uyum (overfitting) yapmasını engeller.
- **Ezberlemeye Karşı Dayanıklıdır**: Daha genelleştirilebilir bir model üretir.

---

### Random Forest (Rastgele Orman)

Random Forest, bagging yönteminin üzerine **random subspace** yöntemi eklenerek geliştirilmiş bir topluluk öğrenme yöntemidir. Birden çok karar ağacı oluşturarak tahminlerin birleştirilmesi prensibiyle çalışır.

**Random Forest’ın Temel Özellikleri:**

- **Bagging ve Random Subspace Yöntemlerinin Birleşimi**: Bagging yöntemiyle bootstrap örnekleme yapılırken, her bir düğümde değişkenler random subspace yöntemiyle seçilir.
- **Rastgele Değişken Seçimi**: Karar ağacının her düğümünde en iyi dallara ayırıcı (bilgi kazancı veya Gini indeksi gibi) değişken, tüm değişkenler yerine rastgele seçilen daha az sayıdaki değişken arasından seçilir.
- **2/3 Veri Kullanımı**: Her bir ağaç, veri setinin yaklaşık %66’sı (2/3’ü) ile eğitilir. Geriye kalan %33’lük veri, ağaçların performansını değerlendirmek ve değişken önemini hesaplamak için kullanılır (Out-of-Bag Error - OOB).
- **Rastgele Alt Küme Seçimi**: 
  - **Regresyon için**: Her düğümde rastgele seçilen değişken sayısı \(p/3\) (toplam değişken sayısının üçte biri).
  - **Sınıflandırma için**: Her düğümde rastgele seçilen değişken sayısı \(\sqrt{p}\) (toplam değişken sayısının karekökü).

**Random Forest’ın Avantajları:**

- **Genel Performans Artışı**: Hem varyansı hem de sapmayı azaltarak daha dengeli ve doğru tahminler yapar.
- **Değişken Önemi Belirleme**: Hangi değişkenlerin modelin performansına en çok katkıda bulunduğunu analiz eder.
- **Aşırı Uyumdan Kaçınma**: Topluluk yöntemleriyle modelin oek daha detaylı ve anlaşılır hale getirdi.

```python
# kütüphaneler

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import mean_squared_error,r2_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from sklearn.preprocessing import StandardScaler
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn import neighbors
from sklearn.svm import SVR
```

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
# Veri setinin yüklenmesi
df = pd.read_csv(r"C:\Users\Emirhan Tozlu\Desktop\machine learning\Doğrusal Olmayan Regresyon Modelleri\KNN ALgoritması\Hitters.csv")
df = df.dropna()  


dms = pd.get_dummies(df[["League", "Division", "NewLeague"]])


y = df["Salary"]
X_ = df.drop(["Salary", "League", "Division", "NewLeague"], axis=1).astype('float64')
X = pd.concat([X_, dms[["League_N", "Division_W", "NewLeague_N"]]], axis=1)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Model & Tahmin
```

```python
rf_model=RandomForestRegressor(random_state=42).fit(X_train,y_train)
```

```python
rf_model
```

```python
y_pred=rf_model.predict(X_test)
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
# Model Tuning
```

```python
rf_params={
    "max_depth": range(5,8,10),
    "max_features":[1,2,5,10],
    "n_estimators":[200,500,1000,2000],
    "min_samples_split":[2,10,80,100]
}
```

```python
rf_cv_model=GridSearchCV(rf_model,rf_params,cv=10,n_jobs=-1,verbose=2).fit(X_train,y_train)
```

```python
rf_cv_model.best_params_
```

```python
rf_model=RandomForestRegressor(
    random_state=42,
    max_depth=5,
    max_features=2,
    min_samples_split=2,
    n_estimators=2000)

rf_tuned=rf_model.fit(X_train,y_train)
```

```python
y_pred=rf_tuned.predict(X_test)
np.sqrt(mean_squared_error(y_pred,y_test))
```

### Değişken Önem Düzeyi

```python
pd.DataFrame({"Importance":rf_tuned.feature_importances_*100})
```

```python
Importance = pd.DataFrame({"Importance":rf_tuned.feature_importances_*100},index=X_train.columns)

Importance.sort_values(by="Importance",
                      axis=0,
                      ascending=True).plot(kind="barh",color="r")

plt.xlabel('Veriable Importance')
plt.gca().legend_=None
```
