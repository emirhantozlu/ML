### Decision Tree Regressor ve CART Algoritması: Genel Bir Bakış

**CART (Classification and Regression Tree)** algoritması, hem sınıflandırma hem de regresyon problemleri için kullanılan, veri setlerindeki karmaşık yapıları basit ve anlaşılır karar ağaçlarına dönüştürmeyi hedefleyen güçlü bir yöntemdir. Scikit-learn kütüphanesinde yer alan `DecisionTreeRegressor`, CART algoritmasını kullanarak regresyon problemlerinde tahmin yapar. İşte bu algoritma ve ilgili parametreleri hakkında genel bir özet:

---

### **CART Algoritmasının Hedefi ve İşleyişi**

**Hedef**:  
CART, heterojen veri setlerini hedef değişkene göre daha homojen alt gruplara ayırır. Bu süreç, bir dizi karar noktası (dallanma) belirleyerek veri setini alt gruplara bölmeyi içerir. Sınıflandırma problemlerinde **Gini İndeksi** veya **Entropi**, regresyon problemlerinde ise **Karesel Hata** (Mean Squared Error) gibi metrikler kullanılır.

**İşleyiş**:
1. **Dallanma (Splitting)**: Her düğümde, veri setini en iyi şekilde ayıran özellik ve eşik değeri seçilir.
2. **Yaprak Düğüm (Leaf Node)**: Daha fazla dallanmanın yapılmadığı nihai grupları oluşturur. Bu düğümler homojen alt grupları temsil eder.
3. **Budama (Pruning)**: Ağaç karmaşıklığını azaltarak aşırı öğrenmeyi (overfitting) engeller.

---

### **Scikit-learn `DecisionTreeRegressor` Parametreleri**

`DecisionTreeRegressor` modelini özelleştirmek için kullanılan başlıca parametreler:

- **criterion**: Bölme kalitesini ölçmek için kullanılan metrik. Varsayılan olarak `squared_error` (Mean Squared Error) kullanılır. 
- **max_depth**: Ağacın maksimum derinliği. Aşırı öğrenmeyi engellemek için bu değerin sınırlandırılması önemlidir.
- **min_samples_split**: Bir düğümün bölünebilmesi için gereken minimum örnek sayısı.
- **min_samples_leaf**: Yaprak düğümde bulunması gereken minimum örnek sayısı. Modelin genelleştirilmesine yardımcı olur.
- **max_features**: Bölme sırasında değerlendirilecek maksimum özellik sayısı. Varsayılan olarak tüm özellikler kullanılır.
- **random_state**: Rastgelelik unsurlarını kontrol ederek sonuçların tekrarlanabilirliğini sağlar.
- **ccp_alpha**: Maliyet karmaşıklığı budama için bir parametre. Daha küçük bir `ccp_alpha` değeri daha büyük bir ağaç üretirken, daha büyük bir değer ağacı küçültür.

---

### **Avantajlar ve Dezavantajlar**

**Avantajlar**:
- **Yorumlanabilirlik**: Ağaç yapısı kolayca görselleştirilebilir ve anlaşılabilir.
- **Esneklik**: Hem sınıflandırma hem de regresyon problemlerinde kullanılabilir.
- **Ön işleme ihtiyacının azlığı**: Ölçekleme veya dönüşüm gibi veri ön işleme adımlarına genellikle ihtiyaç duymaz.

**Dezavantajlar**:
- **Aşırı öğrenme eğilimi**: Model, veri setine çok fazla uyum sağlayabilir, bu da genelleştirme performansını düşürebilir.
- **Performans**: Büyük ve karmaşık veri setlerinde, özellikle Random Forest veya Gradient Boosting gibi topluluk yöntemlerine kıyasla daha düşük tahmin başarısı gösterebilir.
- **Parametre Ayarı**: Doğru ayarların yapılması (budama, derinlik sınırı, vb.) zaman alabilir ve dikkat gerektirir.

---

### **Özet**

CART algoritması, karar ağaçları üzerinden veri setlerindeki yapıyı modellemek için güçlü bir araçtır. `DecisionTreeRegressor`, bu algoritmanın regresyon problemleri için uygulanmış halidir. Doğru parametre ayarları ve budama teknikleriyle güçlü ve anlaşılır modeller oluşturulabilir. Ancak büyük veri setleri ve karmaşık problemler için daha gelişmiş topluluk yöntemleriyle birleştirilmesi gerekebilir.

```python
# kütüphaneler

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import mean_squared_error,r2_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from sklearn.preprocessing import StandardScaler
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn import neighbors
from sklearn.svm import SVR
```

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
# Veri setinin yüklenmesi
df = pd.read_csv(r"C:\Users\Emirhan Tozlu\Desktop\machine learning\Doğrusal Olmayan Regresyon Modelleri\KNN ALgoritması\Hitters.csv")
df = df.dropna()  


dms = pd.get_dummies(df[["League", "Division", "NewLeague"]])


y = df["Salary"]
X_ = df.drop(["Salary", "League", "Division", "NewLeague"], axis=1).astype('float64')
X = pd.concat([X_, dms[["League_N", "Division_W", "NewLeague_N"]]], axis=1)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train= pd.DataFrame(X_train["Hits"])
X_test = pd.DataFrame(X_test["Hits"])
```

```python
# Model & Tahmin
cart_model=DecisionTreeRegressor(max_leaf_nodes = 3 ) 
```

```python
cart_model.fit(X_train,y_train)
```

```python
X_grid=np.arange(min(np.array(X_train)),max(np.array(X_train)), 0.01)
X_grid=X_grid.reshape(len(X_grid),1)

plt.scatter(X_train,y_train,color="red")

plt.plot(X_grid,cart_model.predict(X_grid),color='blue')

plt.title("CART Regresyon Ağacı")
plt.xlabel("Atış Sayısı (Hits)")
plt.ylabel("Maaş (Salary)");
```

```python
# Tek Değişkenle Tahmin 
```

```python
cart_model.predict(X_test)[0:5]
```

```python
y_pred=cart_model.predict(X_test)
np.sqrt(mean_squared_error(y_test,y_pred))
```
