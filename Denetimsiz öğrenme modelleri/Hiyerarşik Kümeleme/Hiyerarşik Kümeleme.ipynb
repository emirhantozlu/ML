# **Hiyerarşik Kümeleme**  
Hiyerarşik kümeleme, gözlem birimlerini birbirlerine olan benzerliklerine göre alt kümelere ayırmayı amaçlayan bir kümeleme yöntemidir. Bu yöntem, gözlemleri daha detaylı alt kümelere ayırmak istediğimiz durumlarda tercih edilir. Hiyerarşik kümeleme, veri setini hem belirli bir sayıda kümeye bölebilir hem de bu kümeleri kendi içinde alt kümelere ayırma imkânı sunar.  

---

### Hiyerarşik Kümeleme Türleri:
1. **Agglomerative (Birleştirici) Yaklaşım:**  
   - Başlangıçta, her gözlem birimi kendi başına bir küme olarak kabul edilir (gözlem sayısı kadar küme vardır).
   - **Adım 1:** Veri setindeki birbirine en yakın iki gözlem belirlenir.  
   - **Adım 2:** Bu iki gözlem birleştirilerek yeni bir küme oluşturulur. Artık veri seti, bu birleşimle oluşan yeni kümelerden oluşur.  
   - **Adım 3:** Oluşan yeni kümeler de kendi aralarındaki benzerliklere göre tekrar birleştirilir. Bu süreç, tüm gözlemler tek bir küme haline gelene kadar devam eder.  

   Birleştirici yöntemde, gözlemler arasındaki yakınlık veya benzerlik, çeşitli uzaklık ölçütleriyle belirlenir. Örneğin:
   - **Öklid Uzaklığı**  
   - **Manhattan Uzaklığı**  
   - **Korelasyon Katsayısı**

2. **Divisive (Bölümleyici) Yaklaşım:**  
   - Başlangıçta, tüm veri seti tek bir küme olarak ele alınır.  
   - **Adım 1:** Tüm gözlemlerin bir arada bulunduğu küme, iki alt kümeye ayrılır.  
   - **Adım 2:** Oluşan alt kümeler de kendi içlerinde birbirine benzemeyen alt kümelere bölünür.  
   - **Adım 3:** Bu süreç, her gözlem biriminin kendi başına bir küme haline gelmesine kadar devam eder.  

---

### Hiyerarşik ve Diğer Kümeleme Yöntemlerinin Karşılaştırılması:
1. **Hiyerarşik Kümeleme:**  
   - Küme sayısına, **dendrogram** adı verilen bir grafik yardımıyla karar verilir.  
   - Hem gözlem birimleri (veri noktaları) hem de değişkenler bazında kümeleme yapılabilir.  

2. **Hiyerarşik Olmayan (Örneğin K-Means):**  
   - Küme sayısı, uygulamadan önce kullanıcı tarafından belirlenmelidir.  
   - Yalnızca gözlem birimlerinin kümelenmesi mümkündür, değişkenlerin kümelenmesi yapılamaz.  

---

### Karar Ağaçları ile Hiyerarşik Kümeleme Farkları:
1. **Hedef Değişken:**  
   - **Karar Ağaçları:** Kümeleme işlemi, bir hedef değişkene (bağımlı değişken) göre yapılır.  
   - **Hiyerarşik Kümeleme:** Hedef değişken yoktur; gözlemler, bağımsız değişkenler üzerinden yapılan uzaklık hesaplamalarına göre kümelere ayrılır.  

2. **Amaç:**  
   - **Karar Ağaçları:** Sınıflandırma veya regresyon işlemi yapmayı hedefler.  
   - **Hiyerarşik Kümeleme:** Gözlemleri benzerliklerine göre gruplandırmayı hedefler.  

3. **Yapı:**  
   - **Karar Ağaçları:** Hedef değişkenin etkisiyle dallanarak ilerleyen bir yapı oluşturur.  
   - **Hiyerarşik Kümeleme:** Veri setini tamamen bağımsız değişkenlere dayalı bir şekilde alt kümelere böler veya birleştirir.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
```

```python
df=pd.read_csv(r"C:\Users\Emirhan Tozlu\Desktop\machine learning\Denetimsiz öğrenme modelleri\Veri seti\USArrests.csv", index_col=0)
df.head()
```

```python
hc_complete=linkage(df,"complete")
hc_average=linkage(df,"average")
```

```python
plt.figure(figsize= (15,10))
plt.title("Hiyerarşik Kümeleme Dendrogramı")
plt.xlabel("Gözlem birimleri")
plt.ylabel("uzaklıkları")
dendrogram(hc_complete,leaf_font_size=10);
```

```python
plt.figure(figsize= (15,10))
plt.title("Hiyerarşik Kümeleme Dendrogramı")
plt.xlabel("Gözlem birimleri")
plt.ylabel("uzaklıkları")
dendrogram(hc_complete,
           truncate_mode="lastp",
           p=4,
           show_contracted=True,
           leaf_font_size=10);
```

```python
plt.figure(figsize= (15,10))
plt.title("Hiyerarşik Kümeleme Dendrogramı")
plt.xlabel("Gözlem birimleri")
plt.ylabel("uzaklıkları")
dendrogram(hc_complete,
           truncate_mode="lastp",
           p=10,
           show_contracted=True,
           leaf_font_size=10);
```
