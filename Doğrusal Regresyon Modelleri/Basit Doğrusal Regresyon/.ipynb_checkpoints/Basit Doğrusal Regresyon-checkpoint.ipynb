# Basit Doğrusal Regresyon

Temel amaç, bağımlı ve bağımsız değişken arasındaki ilişkiyi ifade eden doğrusal fonksiyonub bulmaktır.  

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv(r"C:\Users\Emirhan Tozlu\Desktop\machine learning\Doğrusal Regresyon Modelleri\Basit Doğrusal Regresyon\Advertising.csv")
df=df.iloc[:,1:len(df)]
df.head()
```

```python
df.info()
```

```python
import seaborn as sns
sns.jointplot(x="TV",y="sales",data=df,kind="reg");
```

### Model Kurma

```python
from sklearn.linear_model import LinearRegression
X = df[["TV"]]
X.head()
```

```python
y=df[["sales"]]
```

```python
reg= LinearRegression() # linear regresyon model nesnesi oluşturdum.
```

```python
model=reg.fit(X,y)
```

```python
model
```

```python
dir(model)
```

```python
model.intercept_ # beta0
```

```python
model.coef_ # beta1
```

```python
# rkare
model.score(X,y) # model score'umuz. bu şu anlama geliyor. bağımsız değişkenimizin bağımlı değişkenin 
# yüzde kaçını açıklayabildiğini gösterir.
```

### Tahmin

```python
g = sns.regplot(x=df["TV"], y=df["sales"], ci=None, scatter_kws={'color': 'r', 's': 9})
g.set_title("Model Denklemi: sales = 7.03 + TV*0.05")
g.set_ylabel("Satış Sayısı")
g.set_xlabel("TV Harcamaları")
plt.xlim(-10, 310)
plt.ylim(bottom=0)
plt.show()
```

sns.regplot(), iki değişken arasında regresyon çizgisi ile bir dağılım grafiği çizer.
- x=df["TV"]: TV harcamalarını x eksenine yerleştirir.
- y=df["sales"]: Satışları y eksenine yerleştirir.
- ci=None: Güven aralığını göstermez.
- scatter_kws={'color': 'r', 's': 9}: Dağılım noktalarını kırmızı renkte ve boyutunu 9 olacak şekilde   ayarlar.

### Artıklar Ve Makine Öğrenmesindeki Önemi

MSE: Hata Kareler Ortalaması

RMSE: Hata Kareler Ortalaması Karekökü

```python
y.head()
```

```python
model.predict(X)[0:6]
```

```python
gercek_y=y[0:10]
```

```python
tahmin_y=pd.DataFrame(model.predict(X)[0:10])
```

```python
hatalar= pd.concat([gercek_y,tahmin_y],axis=1)
hatalar.columns=["gercek_y","tahmin_y"]
hatalar
```

```python
hatalar["hata"] = hatalar["gercek_y"] - hatalar["tahmin_y"]
```

```python
hatalar
```

```python
hatalar["hata_kareler"]=hatalar["hata"]**2
hatalar
```

```python
hatalar["hata_kareler"].mean()
```
